# 🧬 Pangenome Analysis Pipeline 
## Part 2: Genome Annotation using Prokka

### 🎯 Objective
To construct a pangenome, all genomes must be annotated in a consistent and standardized way. 
We will use **Prokka** to rapidly annotate our 50 `FASTA` files. Prokka identifies coding sequences (CDS), genes, and generates the necessary `.gff` (General Feature Format) files required by downstream tools like **Roary**.

### 🗺️ Workflow:
1. Create a master directory (`annotated_genomes/`) for all outputs.
2. Loop through each `.fna` file in the `my_genomes/` folder.
3. Execute `prokka` iteratively, utilizing multiple CPUs for efficiency.
4. Validate the completion by counting the generated `.gff` files.

```python
%%bash

# 1. Create a main directory for all Prokka outputs
mkdir -p annotated_genomes

# 2. Define the organism (Change these if your target is different)
GENUS="Acinetobacter"
SPECIES="baumannii"

echo "🚀 Starting Prokka annotation for 50 genomes ($GENUS $SPECIES)..."
echo "⏳ This step is computationally intensive and might take a while. Please wait."

# 3. Loop through each FASTA file
for fasta in my_genomes/*.fna; do
    filename=$(basename "$fasta" .fna)
    
    echo "🧬 Annotating: $filename"
    
    # Run Prokka with Genus and Species specific databases
    prokka --cpus 4 --genus "$GENUS" --species "$SPECIES" --outdir "annotated_genomes/$filename" --prefix "$filename" "$fasta" --force --quiet
done

# 4. Verify the final GFF files
echo "✅ Annotation complete! Counting the generated GFF files:"
find annotated_genomes -name "*.gff" | wc -l
```
